```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

# A. Una función que calcule el precio total de la compra
precios=[]
for key in compra: 
    precios.append(articulos[key]) 

print(precios)

precio_compra= reduce(lambda x,y: x+y, precios, 0)
print('Precio total de la compra (sin IVA y sin descuento): ', precio_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

precio_descuento=reduce(lambda x,y: x+y, list(map(lambda z: z-(z*0.10), precios)), 0)
print('Precio con descuento del 10% es: ', precio_descuento)

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
if precio_compra>100: 
    precio_descuento100=reduce(lambda x,y: x+y, list(map(lambda z: z-(z*0.10), precios)), 0)
    print('Tu compra con descuento al superar los 100€ es: ', precio_descuento100)
else: 
    print('Tienes que superar el importe de 100€ para recibir el 10% de descuento')


# D. Una función que calcule el precio total aplicando el IVA (21%)
precio_iva= precio_descuento=reduce(lambda x,y: x+y, list(map(lambda z: z+(z*0.21), precios)), 0)
print('Precio con IVA aplicado', precio_iva)

# E.lista los artículos cuyo precio es superior a 20€

articulos_mayor20=[]
for key, value in articulos.items(): 
    if value>20: 
        articulos_mayor20.append(key)
articulos_mayor20

```
